```python
import pandas as pd
import numpy as np
```

```python
def mod_false(df):
    for i in range(len(df)):
        df.loc[i,'modifications'] = f"{df.loc[i, 'N-terminus']}, {df.loc[i, 'C-terminus']}"
        df.loc[i,'is_mod'] = 'True'
    return df
```

```python
df = pd.read_excel('../raw/PepTherDia.xlsx')
```

```python
df2 = df[['Sequence', 'N-terminus', 'C-terminus', 'Terminal Half-Life']]
mod_false(df2)
df2.drop(['N-terminus', 'C-terminus'], axis=1, inplace=True)
df2.rename(columns={'Sequence': 'sequence', 'Terminal Half-Life': 'half_life_seconds'}, inplace=True)
```

```python
df2
```

```python
df_filtered = df2[df2['half_life_seconds'].str.contains(r'^\d+\.?\d*\s*(hours|min|minutes)$')]
df_filtered.reset_index(drop=True, inplace=True)
```

```python
df_filtered
```

```python
for i in range(len(df_filtered)):
    if 'hours' in df_filtered.loc[i, 'half_life_seconds']:
        df_filtered.loc[i, 'half_life_seconds'] = float(df_filtered.loc[i, 'half_life_seconds'].split(' ')[0]) * 3600
    else:
        df_filtered.loc[i, 'half_life_seconds'] = float(df_filtered.loc[i, 'half_life_seconds'].split(' ')[0]) * 60
```

```python
df_filtered
```

```python
#Por si se quieren quitar las modificaciones
df_filter = df_filtered[df_filtered['is_mod'] == 'False']
df_filter = df_filtered.drop(columns=['is_mod', 'modifications'])
```
