```python
import pandas as pd
import sys
```

```python
folder='../datasets'
list_null = []
```

```python
df = pd.read_csv("../datasets/PEP_LIFE.csv")
#se crea un dataframe agrupando los datos duplicados, guardando sus tiempos de vida, si este ya ha sido procesado anteriormente, se sale del programa
#Sort mantiene el orden de las secuencias original y agg(list) agrupa los tiempos de vida en una lista
```

if (str(df.iloc[0].values[1])).startswith('[')==False:
    df_grouped = df.groupby('sequence', sort=False).agg(list)
else:
    print("El archivo ya ha sido procesado anteriormente. Salida del programa.")
    sys.exit()

```python
for i in range(0, len(df)):
    sequence = df.iloc[i].values[0]
    time = df.iloc[i].values[1]
    print(sequence, time)
    if pd.isna(time):
        list_null.append(time)
```

```python
print(list_null)
```

```python
#Itera sobre las secuencias y tiempos de vida, si hay algun tiempo de vida nulo lo añade a la lista
for i in range(0, len(df_grouped)):
    sequence = df_grouped.index[i]
    time = df_grouped.iloc[i].values[0]
    if pd.isna(time).any():
        list_null.append({"sequence": sequence, "time": time})
list_null= pd.DataFrame(list_null, columns=['sequence', 'time'])
```

```python
list_null
```

```python
for i in range(0, len(list_null)):
    sequence = list_null.iloc[i].values[0]
    time = list_null.iloc[i].values[1]
    if time == 'nan':
        print("La secuencia", sequence, "no tiene tiempo de vida.")
```
