 Función para intentar convertir a float y devolver False si falla
def to_float_or_false(value):
    try:
        return float(value)
    except ValueError:
        return False

# Aplicar la función a la columna 'half_life_seconds'
df['half_life_seconds'] = df['half_life_seconds'].apply(to_float_or_false)
